```python
def is_odd(a):
    if a % 2 != 0:
        return True
    else:
        False

def is_more_or_equal_5(a):
    if a >= 5:
        return True
    else:
        return False

while 1:
    a = int(input('How many games do you want to play? (Minimum 5 games): '))
    if is_odd(a) and is_more_or_equal_5(a):
        print("Input taken successfully")
        print("Number is odd and greater than five")
        print("The game starts now:")
        games_played=0
        number_of_games=a
        played=0
        machine_wins=0
        player_wins=0
        to_win=(number_of_games/2)+1
        left=number_of_games
        while played<number_of_games:
            if player_wins<to_win or machine_wins<to_win:
                import sys
                import random
                random = random.randrange(0, 5)
                print("1)rock")
                print("2)paper")
                print("3)scissors")
                print("4)lizzard")
                print("5)spock")
                if random == 0:
                    machine = "rock"
                elif random == 1:
                    machine = "paper"
                elif random == 2:
                    machine = "scissors"
                elif random == 3:
                    machine = "lizzard"
                elif random == 4:
                    machine = "spock"
                option = int(input("What do you want to use?: "))
                while option > 5 or option==0:
                    print("Not a correct option.")# pick again
                    option = int(input("What do you want to use?: "))
                if option == 1:
                    player = "rock"
                elif option == 2:
                    player = "paper"
                elif option == 3:
                    player = "scissors"
                elif option == 4:
                    player = "lizzard"
                elif option == 5:
                    player = "spock"
                print("You choose: ", player)
                print("The machine has chosen: ", machine)

                if machine == "rock" and player == "paper":
                    print("Ganaste, paper envuelve rock")
                    player_wins+=1
                elif machine == "paper" and player == "scissors":
                    print("Ganaste, scissors corta paper")
                    player_wins+=1
                elif machine == "scissors" and player == "rock":
                    print("Ganaste, rock pisa scissors")
                    player_wins+=1
                elif machine == "paper" and player == "rock":
                    print("perdiste, paper envulve rock")
                    machine_wins+=1
                elif machine == "scissors" and player == "paper":
                    print("perdiste, scissors corta paper")
                    machine_wins+=1
                elif machine == "rock" and player == "scissors":
                    print("perdiste, rock pisa scissors")
                    machine_wins+=1
                elif machine == "spock" and player == "scissors":
                    print("perdiste, spock rompe scissors")
                    machine_wins+=1
                elif machine == "scissors" and player == "spock":
                    print("Ganaste, spock rompe scissors")
                    player_wins+=1
                elif machine == "spock" and player == "rock":
                    print("perdiste, spock vaporiza rock")
                    machine_wins+=1
                elif machine == "rock" and player == "spock":
                    print("Ganaste, spock vaporiza scissors")
                    player_wins+=1
                elif machine == "spock" and player == "paper":
                    print("Ganaste, paper desautoriza spock")
                    player_wins+=1
                elif machine == "paper" and player == "spock":
                    print("perdiste, paper desautoriza spock")
                    machine_wins+=1
                elif machine == "spock" and player == "lizzard":
                    print("Ganaste, lizzard envenena a spock")
                    player_wins+=1
                elif machine == "lizzard" and player == "spock":
                    print("perdiste, lizzard envenena a spock")
                    machine_wins+=1  
                elif machine == "lizzard" and player == "rock":
                    print("Ganaste, rock aplasta lizzard")
                    player_wins+=1
                elif machine == "rock" and player == "lizzard":
                    print("perdiste, rock aplasta lizzard")
                    machine_wins+=1
                elif machine == "lizzard" and player == "paper":
                    print("perdiste, lizzard devora paper")
                    machine_wins+=1
                elif machine == "paper" and player == "lizzard":
                    print("Ganaste, lizzard devora paper")
                    player_wins+=1
                elif machine == "lizzard" and player == "scissors":
                    print("Ganaste, scissors decapita lizzard")
                    player_wins+=1
                elif machine == "scissors" and player == "lizzard":
                    print("perdiste, scissors decapita lizzard")
                    machine_wins+=1
                elif machine == player:
                    print("tie")    
                print("There are ",left,"games to play\n")               
                print("Player has won",player_wins ,"game")
                print("The machine has won",machine_wins ,"game")
                played+=1
                left-=1
            else:
                print("hay un ganador")
        if player_wins<machine_wins:
            print("The machine is the winner!!!! With ",machine_wins ,"game wins")
        elif machine_wins==to_win:
            print("The machine is the winner!!!! With ",machine_wins ,"game wins")
        elif player_wins==to_win:
            print("Player is the winner!!!! With ",player_wins ,"game wins")           
        elif player_wins==machine_wins:
            print("There are no winners, TIE!!!!")
        else:
            print("Player is the winner!!!! With ",player_wins ,"game wins")
        sys.exit() ## I have put it so that the infinite loop ends when the number of games that were wanted to be played is reached
    else:
        if not is_odd(a):
            print("failed! The number of games must be odd")
            continue
        if not is_more_or_equal_5(a):
            print("failed! The number of games must be greater than or equal to 5")
            continue
```
